```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt; 
%matplotlib inline
```

```
def gaji_kecil(x):
    L=1 #tinggi maks
    k=10 #seberapa belok
    a=0.5 #titik tengah
    return L/(1+np.exp(-k*(-x+a)))
        
def gaji_lumayan(x):
    a=1 #titik tengah
    w=0.5 #lebar
    return np.exp(-((x-a)/w)**2)
  
def gaji_lebih(x):
    L=1 #tinggi maks
    k=10 #seberapa belok
    a=1.5 #titik tengah
    return L/(1+np.exp(-k*(x-a)))
```

```
volume = np.arange(0,2, 0.01)
plt.plot(volume, gaji_kecil(volume), color='green', label='Kecil')
plt.plot(volume, gaji_lumayan(volume), color='blue', label='Lumayan')
plt.plot(volume, gaji_lebih(volume), color='red', label='Lebih')

plt.legend()
plt.show()
```

```
volume = np.arange(0,90, 0.1)
plt.plot(volume, hutang_kecil(volume), color='green', label='Kecil')
plt.plot(volume, hutang_lumayan(volume), color='blue', label='Lumayan')
plt.plot(volume, hutang_lebih(volume), color='red', label='Lebih')

plt.legend()
plt.show()
```
