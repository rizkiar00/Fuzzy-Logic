```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt; 
from google.colab import files
%matplotlib inline
!git clone https://github.com/rizkiar00/Fuzzy-Logic
```

```
def gaji_kecil(x):
    L=1 #tinggi maks
    k=10 #seberapa belok
    a=0.5 #titik tengah
    return L/(1+np.exp(-k*(-x+a)))
        
def gaji_lumayan(x):
    a=1 #titik tengah
    w=0.2 #lebar
    return np.exp(-((x-a)/w)**2)
  
def gaji_lebih(x):
    L=1 #tinggi maks
    k=10 #seberapa belok
    a=1.5 #titik tengah
    return L/(1+np.exp(-k*(x-a)))
```

```
volume = np.arange(0,2, 0.01)
plt.plot(volume, gaji_kecil(volume), color='green', label='Kecil')
plt.plot(volume, gaji_lumayan(volume), color='blue', label='Lumayan')
plt.plot(volume, gaji_lebih(volume), color='red', label='Lebih')

plt.legend()
plt.title('Gaji')
plt.show()
```

```
def hutang_kecil(x):
    L=1 #tinggi maks
    k=0.25 #seberapa belok
    a=22.5 #titik tengah
    return L/(1+np.exp(-k*(-x+a)))
        
def hutang_lumayan(x):
    L = 1 #tinggi maks
    a=45 #titik tengah
    w=10 #lebar
    return L*np.exp(-((x-a)/w)**2)
  
def hutang_lebih(x):
    L=1 #tinggi maks
    k=0.25 #seberapa belok
    a=66.5 #titik tengah
    return L/(1+np.exp(-k*(x-a)))
```

```
volume = np.arange(0,90, 0.1)
plt.plot(volume, hutang_kecil(volume), color='green', label='Kecil')
plt.plot(volume, hutang_lumayan(volume), color='blue', label='Lumayan')
plt.plot(volume, hutang_lebih(volume), color='red', label='Lebih')

plt.legend()
plt.title('Hutang')
plt.show()
```

```
def harta_kecil(x):
    L=1 #tinggi maks
    k=20 #seberapa belok
    a=0.25 #titik tengah
    return L/(1+np.exp(-k*(-x+a)))
        
def harta_lumayan(x):
    L = 1 #tinggi maks
    a=0.5 #titik tengah
    w=0.1 #lebar
    return L*np.exp(-((x-a)/w)**2)
  
def harta_lebih(x):
    L=1 #tinggi maks
    k=20 #seberapa belok
    a=0.75 #titik tengah
    return L/(1+np.exp(-k*(x-a)))
```

```
volume = np.arange(0,1, 0.01)
plt.plot(volume, harta_kecil(volume), color='green', label='Kecil')
plt.plot(volume, harta_lumayan(volume), color='blue', label='Lumayan')
plt.plot(volume, harta_lebih(volume), color='red', label='Lebih')

plt.legend()
plt.title('Harta')
plt.show()
```

```
def rule(gaji,g, hutang,h):
    if gaji == 'kecil' and hutang == 'kecil':
        return (np.minimum(g,h)) #lumayan
    elif gaji == 'kecil' and hutang == 'lumayan':
        return (np.minimum(g,h)) #kecil
    elif gaji == 'kecil' and hutang == 'lebih':
        return (np.minimum(g,h)) #kecil
    elif gaji == 'lumayan' and hutang == 'kecil':
        return (np.minimum(g,h)) #lebih
    elif gaji == 'lumayan' and hutang == 'lumayan':
        return (np.minimum(g,h)) #lumayan
    elif gaji == 'lumayan' and hutang == 'lebih':
        return (np.minimum(g,h)) #kecil
    elif gaji == 'lebih' and hutang == 'kecil':
        return (np.minimum(g,h)) #lebih
    elif gaji == 'lebih' and hutang == 'lumayan':
        return (np.minimum(g,h)) #lebih
    elif gaji == 'lebih' and hutang == 'lebih':
        return (np.minimum(g,h)) #lumayan
```

```
data = pd.read_csv('Fuzzy-Logic/DataTugas2.csv',skipinitialspace=True)
del data['No']
data
```

```
gkecil = []
glumayan = []
glebih = []
hkecil = []
hlumayan = []
hbesar = []
for index, row in data.iterrows():
    gkecil.append(gaji_kecil(row['Pendapatan']))
    glumayan.append(gaji_lumayan(row['Pendapatan']))
    glebih.append(gaji_lebih(row['Pendapatan']))
    hkecil.append(hutang_kecil(row['Hutang']))
    hlumayan.append(hutang_lumayan(row['Hutang']))
    hbesar.append(hutang_lebih(row['Hutang']))
#     fuzzy_output.append(data)
d = {'gkecil': gkecil, 'glumayan': glumayan, 'glebih':glebih, 'hkecil':hkecil, 'hlumayan':hlumayan, 'hbesar':hbesar}
fuzzy_output = pd.DataFrame(data=d)
fuzzy_output = fuzzy_output[['gkecil', 'glumayan', 'glebih', 'hkecil', 'hlumayan', 'hbesar']]
fuzzy_output
```

```
hakecil = []
halumayan = []
halebih = []
for index, row in fuzzy_output.iterrows():
    hakecil.append(max(rule('kecil',row['gkecil'],'lumayan',row['hlumayan']),
                       rule('lumayan',row['glumayan'],'lebih',row['glebih']),
                       rule('kecil',row['gkecil'],'lebih',row['glebih'])))
    halumayan.append(max(rule('kecil',row['gkecil'],'kecil',row['hkecil']),
                         rule('lumayan',row['glumayan'],'lumayan',row['hlumayan']),
                         rule('lebih',row['glebih'],'lebih',row['glebih'])))
    halebih.append(max(rule('lebih',row['glebih'],'lumayan',row['hlumayan']),
                       rule('lebih',row['glebih'],'kecil',row['hkecil']),
                       rule('lumayan',row['glumayan'],'kecil',row['gkecil'])))
d = {'hakecil':hakecil, 'halumayan':halumayan, 'halebih':halebih}
output = pd.DataFrame(data=d)
output = output[['hakecil', 'halumayan', 'halebih']]
output
```

```
nilaiharta = []
nomor = []
for index, row in output.iterrows():
  nilaiharta.append((1*row['hakecil']+2*row['halumayan']+3*row['halebih'])/(row['hakecil']+row['halumayan']+row['halebih']))
d = {'nilaiharta':nilaiharta}
hasil_defuzifikasi = pd.DataFrame(data=d)
urut = hasil_defuzifikasi.sort_values(by=['nilaiharta'])
urut
```
